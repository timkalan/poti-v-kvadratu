# Vizualizacija v kvadratu

V tem zvezku so demonstitrane nekatere funkcionalnosti, ki jih omogoča razred `Kvadrat`, 
najden v datoteki `kvadrat.py`

```python
import random
import matplotlib.pyplot as plt

from kvadrat import razdalja, Kvadrat

# spreminjanje velikosti grafov
plt.rcParams['figure.figsize'] = [10, 10]    # če grafi preveliki, pomanjšaj ti dve števili
```

## Osnove

`Kvadrat` zna narisati vse svoje točke.

```python
# osnovne funkcionalnosti prikažemo na Kvadratu 'demo'
demo = Kvadrat(100, 0.3)

demo.narisi_tocke()
```

`Kvadrat` zna narisati točke, ki so "dovolj blizu" dani točki.

```python
# najprej narišemo vse
demo.narisi_tocke()

# bližnje točke so potem označene z drugo barvo
# točke izbiramo kot zaporedne številke 0, ..., n-1
demo.narisi_bliznje(tocka=0)
```

`Kvadrat` zna povezati vse bližnje točke s točko - korenom.

```python
# najprej narišemo vse
demo.narisi_tocke()

# podobno kot zgoraj lahko izbiramo točke
demo.narisi_graf_tocke(tocka=0)
```

`Kvadrat` zna seveda tudi narisati celoten graf, ki ga definirajo pravila - večkrat zaporedoma 
nariše graf točke.

```python
# najprej narišemo točke
demo.narisi_tocke()

# dodamo povezave
demo.narisi_graf()
```

`Kvadrat` zna tudi uporabiti Dijkstrov algoritem, da dobi drevo najkrajših poti 
od nekega korena

```python
# najprej narišemo naš graf
demo.narisi_tocke()
demo.narisi_graf()

# dodamo drevo najkrajših poti - označi se tudi koren
# koren izbiramo podobno kot 'tocka' v prejšnjih primerih
demo.narisi_drevo(koren=0)
```

## Nekaj različnih primerov parametrov

Parametri močno vplivajo na izgled in čas izvajanja našega programa.

```python
malo_kratke = Kvadrat(20, 0.1)
malo_dolge = Kvadrat(20, 0.5)

plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
malo_kratke.narisi_tocke()
malo_kratke.narisi_graf()
malo_kratke.narisi_drevo()

plt.subplot(1, 2, 2)
malo_dolge.narisi_tocke()
malo_dolge.narisi_graf()
malo_dolge.narisi_drevo()
```
